## Kmeans standard scaler to non boolean columns 

```python
df = df_prescaled.copy()
```

```python
column_names = list(df.iloc[0,:].reset_index()['index'])
```

```python
columns_to_scale = ['Year_Birth',
 'Income',
 'Kidhome',
 'Teenhome',
 'Dt_Customer',
 'Recency',
 'MntWines',
 'MntFruits',
 'MntMeatProducts',
 'MntFishProducts',
 'MntSweetProducts',
 'MntGoldProds',
 'NumDealsPurchases',
 'NumWebPurchases',
 'NumCatalogPurchases',
 'NumStorePurchases',
 'NumWebVisitsMonth',
 #'AcceptedCmp3',
 #'AcceptedCmp4',
 #'AcceptedCmp5',
 #'AcceptedCmp1',
 #'AcceptedCmp2',
 #'Complain',
 #'Response',
 #'Education_Basic',
 #'Education_Graduation',
 #'Education_Master',
 #'Education_PhD',
 #'Marital_Status_Married',
 #'Marital_Status_Single',
 #'Marital_Status_Together',
 #'Marital_Status_Widow'
                   ]
```

```python
columns_not_to_scale = [#'Year_Birth',
 #'Income',
 #'Kidhome',
 #'Teenhome',
 #'Dt_Customer',
 #'Recency',
 #'MntWines',
 #'MntFruits',
 #'MntMeatProducts',
 #'MntFishProducts',
 #'MntSweetProducts',
 #'MntGoldProds',
 #'NumDealsPurchases',
 #'NumWebPurchases',
 #'NumCatalogPurchases',
 #'NumStorePurchases',
 #'NumWebVisitsMonth',
 'AcceptedCmp3',
 'AcceptedCmp4',
 'AcceptedCmp5',
 'AcceptedCmp1',
 'AcceptedCmp2',
 'Complain',
 'Response',
 'Education_Basic',
 'Education_Graduation',
 'Education_Master',
 'Education_PhD',
 'Marital_Status_Married',
 'Marital_Status_Single',
 'Marital_Status_Together',
 'Marital_Status_Widow'
                   ]
```

```python
df_to_scale = df[columns_to_scale]
```

```python
df_not_to_scale=df[columns_not_to_scale]
```

```python
scaler = StandardScaler()

df_to_scale = pd.DataFrame(scaler.fit_transform(df_to_scale),columns=columns_to_scale)
```

```python
df_to_scale = df_to_scale.reset_index()
```

```python
df_not_to_scale = df_not_to_scale.reset_index().drop(columns='index').reset_index()
```

```python
df_final = df_to_scale.merge(df_not_to_scale)
df_final.drop(columns='index',inplace=True)
```

```python
# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10), locate_elbow=True)

visualizer.fit(df_final)        # Fit the data to the visualizer
visualizer.poof()        # Draw/show/poof the data
plt.show()
```

```python
sillhouette_list= []
for i in range(2,10):  
    model_kmeans = KMeans(n_clusters=i, random_state=0).fit(df_final)
    visualizer = SilhouetteVisualizer(model_kmeans, colors='yellowbrick')
    visualizer.fit(df_final)
    silhouette_avg = silhouette_score(df_final, model_kmeans.labels_)
    visualizer.poof()
    sillhouette_list.append(silhouette_avg)
```
